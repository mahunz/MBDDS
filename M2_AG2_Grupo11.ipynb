El tipo de variables del dataset es:

1. **Num:** es codigo del dataset. El código es único, consecutivo y entero. Es una **variable cuantitativa de tipo discreto.**

2. **Country:** Es una variable clasificada con palabras que no sigue ningún orden en específico. Es una variable **cualitativa nominal**.

3. **Total Cases** Es el cantidad de casos total de COVID. En un número entero. **variable cuantitativa de tipo discreto.**

4. **New Cases** Es la cantidad de nuevos casos de COVID. En un número entero. **variable cuantitativa de tipo discreto.**

5. **Total Deaths** Es la cantidad total de muertes por COVID. En un número entero. **variable cuantitativa de tipo discreto.**

6. **New Deaths** Es la cantidad de nuevas muertes por COVID. En un número entero. **variable cuantitativa de tipo discreto.**

7. **Total Recovered** Es la cantidad de casos recuperados de COVID. En un número entero. **variable cuantitativa de tipo discreto.**

8. **Active Cases** Es la cantidad de casos Activos de COVID. En un número entero. **variable cuantitativa de tipo discreto.**

9. **Serious Critical** Es la cantidad de casos en estado críticos de COVID. En un número entero. **variable cuantitativa de tipo discreto.**

10. **Total Cases per 1M pop** Es la cantidad total de casos por millon de habitantes. Es una variable **cuantitativa continua.**  Nótese: las variables tiene decimales.

11. **Deaths per 1M pop** Es la cantidad total de muertos por millon de habitantes. Es una variable **cuantitativa continua.**  Nótese: las variables tiene decimales.

12. **Total Tests** Es la cantidad total de test realizados. En un número entero. **variable cuantitativa de tipo discreto.**

13. **Tests per 1M pop** Es la cantidad de test por millon de habitantes. Es una variable **cuantitativa continua.**  Nótese: las variables tiene decimales.

14. **Poblacion** Es la población del país. En un número entero. **variable cuantitativa de tipo discreto.**
